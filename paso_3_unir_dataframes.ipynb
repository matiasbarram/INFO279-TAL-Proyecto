# Finalmente se unen los nuevos topicos con el dataset creado en el paso 1.

```python
from os import listdir
import pandas as pd
from pandasql import sqldf
```

### Se toman todos los archivos que fueron revisandos

```python
dfs = []
for file in listdir('.'):
    # categorias_{idx}_paso2.csv
    if 'categorias' in file and 'paso2' in file:
        print(file)
        dfs.append(pd.read_csv(file)) 
        
df = pd.concat(dfs)[['id','title', 'content', 'category']]
df
```

```python
q="""SELECT category, count(*) FROM df GROUP BY category ORDER BY count(*) DESC;"""
result=sqldf(q)
result
```

```python
final_df_link = pd.read_csv("final_df_link.csv")
final_df_topics = pd.read_csv("final_df_topic.csv")
```

### Y se agregan al dataset final

```python
final_df = pd.concat([final_df_link, final_df_topics]).reset_index().drop(columns=['index'])
```

```python
final_df = pd.concat([final_df, df])
final_df.head(3)
```

```python
q="""SELECT category, count(*) FROM final_df GROUP BY category ORDER BY count(*) DESC;"""
result=sqldf(q)
result
```
